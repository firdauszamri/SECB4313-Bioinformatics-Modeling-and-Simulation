```python
#Import all library needed
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense,BatchNormalization, Dropout
import tensorflow as tf
import matplotlib.pyplot as plt
import os
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

#confusion matrix visualization
import seaborn as sns
from sklearn.metrics import confusion_matrix,classification_report
```

```python
### 3. Insert Exploratory data analysis (EDA) steps to analyze and investigate datasets.
data.head() # display first 5 row data
```

```python
### 4. What is the purpose of the code that sets a list of categorical variables
### in a dataset and then casts those variables to the object data type using the astype() function?

# Converting the data type to object treat certain columns as categorical variables,
# enabling easier manipulation, analysis, and interpretation of the data

catagorialList = ['sex','cp','fbs','restecg','exang','ca','thal']
for item in catagorialList:
    data[item] = data[item].astype('object') #casting to object
```

```python
 ### 5. Create more data by categorical variable into indicator variables using 'get_dummies' function

data = pd.get_dummies(data, drop_first=True)
```

```python
### 6. Explain line 3,4 and 5 and print the shape of x and y

y = data['target'].values
y = y.reshape(y.shape[0],1)
x = data.drop(['target'],axis=1)
## 3 = take target value
## 4 = reshape y array into 1 dimension
## 5 = drop 'target' in x data
```

```python
### 7. Create a simple dataset and demonstrate the normalization code on the simple dataset

data = pd.DataFrame({'A': [10, 20, 30], 'B': [100, 200, 300], 'C': [1000, 2000, 3000]})
print('Original dataset:')
print(data)
minx = np.min(data)
maxx = np.max(data)
data_norm = (data - minx) / (maxx - minx)
print('\nNormalized dataset:')
print(data_norm)
```

```python
### 8. Describe the heart dataset after implementing the min max normalization
#Normalize data (range 0 - 1)
minx = np.min(x)
maxx = np.max(x)
x = (x - minx) / (maxx - minx)
x.head()

#The values in the table have been scaled between 0 and 1 using min-max normalization.
```

```python
### 9. Modify the code to split the dataset into train and test (train 70%, val 20% and test 10%).

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
# re-create train and validation set
x_train, x_val, y_train, y_val  = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
# train 70%, validation 20%, test 10%
print(x_train.shape)
print(x_val.shape)
print(x_test.shape)
```

```python
### 10. What is the purpose of each layer in the neural network created using the Sequential() function with 64, 32, and 1 neurons,
### respectively, and softmax and sigmoid activation functions?

model = Sequential() #Allow us to create model layer by layer
model.add(Dense(64, input_dim=21, activation='softmax')) #Softmax turn number data into probabilities which sum to 1
model.add(Dense(32, activation='softmax'))
model.add(Dense(1, activation='sigmoid')) # produce probability value (number between 0 or 1)
model.summary()

## input layer: Softmax activation function is used to convert raw scores into probabilities.
## It squashes the output scores to a range between 0 and 1, ensuring that the sum of all
## probabilities is equal to 1. This makes it suitable for multi-class classification problems,
## as it provides probabilities for each class.

## hidden layer: Similarly to the first layer, softmax activation is used here to convert
## the raw scores into probabilities. However, since this is a hidden layer, the purpose is
## to introduce non-linearity into the network and enable the model to learn complex patterns
## in the data. Softmax is typically used in the output layer for multi-class classification,
## but using it in hidden layers can still be effective depending on the problem.

## output layer: Sigmoid activation function squashes the output between 0 and 1, representing
## the probability of the input belonging to the positive class (class 1 in binary classification).
##It's commonly used in binary classification tasks as it provides a clear decision boundary
## between the two classes.
```

```python
### 11. This code compiles a neural network model with a mean squared error loss function, the Adam optimizer with a learning rate of 0.01,
### and accuracy as a performance metric. What does each of these components mean, and how do they affect the model training and performance?

model.compile(loss='mse',
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.01, beta_1=0.9, beta_2=0.999,epsilon=1e-07, amsgrad=False,name='Adam'),
              metrics=['acc'])
## In training a neural network, the mean squared error (MSE) loss function calculates how far the model's predictions are from the actual values,
## aiming to minimize this difference, much like measuring the accuracy of basketball shots. The Adam optimizer adjusts how the model learns from
## these errors, dynamically changing the learning rate based on the gradients of the loss function, akin to adjusting the learning pace during
## basketball practice based on how close or far your shots are from the hoop. Lastly, accuracy as a performance metric simply tells us the
## percentage of correct predictions the model makes out of all predictions, similar to how we measure the success rate of making basketball shots
## out of all attempts. Together, these components help guide the model's training process towards making more accurate predictions.
```

```python
# start the model training
output = []
early = EarlyStopping(monitor='val_acc', patience=400, mode='auto')
checkpoint = ModelCheckpoint(model_loc+"heart_disease_best_model.hdf5", monitor='val_acc', verbose=0, save_best_only=True, mode='auto', save_freq='epoch')
reduce_lr = ReduceLROnPlateau(monitor='val_acc', factor=0.01, patience=100, verbose=1, mode='auto', min_lr=0.001)
callbacks_list = [early]

output = model.fit(x_train, y_train,validation_data=(x_val,y_val), epochs=1000, batch_size=16, verbose=1, callbacks=callbacks_list)
```

```python
### 12. What does the plot generated by this code represent?

plt.plot(output.history['acc'])
plt.plot(output.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
#plt.savefig('Accuracy.png',dpi=100) #to save the image
plt.show()
```

```python
### 13. Plot code for the model loss. You can refer to the plot code for model accuracy above.
plt.plot(output.history['loss'])
plt.plot(output.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
#plt.savefig('Loss.png',dpi=100) #to save the image
plt.show()
```

```python
### 14. What is the purpose of evaluating the model on the test dataset?

#model.load_weights(model_loc+"heart_disease_best_model.hdf5")
scores = model.evaluate(x_test, y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
# print("\n%s: %.2f%%" % (model.metrics_names[0], scores[0]))
print("loss:", round(scores[0],2))

## Firstly, it helps us understand how well the model generalizes to unseen data.
## Training a model involves adjusting its parameters to minimize the loss on the training data,
## but we also want to ensure that it can make accurate predictions on new, unseen data.
## By evaluating the model on a separate test dataset, we get an unbiased estimate of its performance.
## Secondly, it helps us compare the performance of different models or different configurations of
## the same model. We can use metrics such as accuracy, loss, precision, recall, etc.,
## to compare how well different models perform on the same test dataset. Finally, it provides insights
## into potential issues such as overfitting, where the model performs well on the training data but
## poorly on new data. If there's a significant difference in performance between the training and
##test datasets, it indicates overfitting, and further adjustments to the model may be needed to improve generalization.
```

```python
#Display detailed prediction
pred = model.predict(x_test)
y = np.round(pred).astype("int16")
idx = 0
ps = 0
fl = 0
for x in pred:
    if y_test[idx]==y[idx]:
        print("\033[30mNo:",idx+1,"Actual:",y_test[idx]," Predicted:",y[idx],"Result: \033[92mPass")
        ps = ps+1
    else:
        print("\033[30mNo:",idx+1,"Actual:",y_test[idx]," Predicted:",y[idx]," Result: \033[91mFail")
        fl = fl+1
    idx = idx + 1
print("\033[30mRight Prediction :",ps, "Wrong Prediction :",fl)
```

```python
### 15. What is Confusion Matrix and why you need it? Explain TP, FP, FN, TN.
### 16. Explain the classification report produce.

y_pred = y
y_true = y_test

cm = confusion_matrix(y_true, y_pred, labels=[0,1])
#cm = confusion_matrix(y_true, y_pred, labels=labels.astype('int'))
f, ax=plt.subplots(figsize=(5,5))
sns.heatmap(cm,annot=True,linewidths=1.5,linecolor="red",fmt=".0f",ax=ax)
plt.xlabel("y_pred")
plt.ylabel("y_true")
plt.show()
print()
print(classification_report(y_true, y_pred, labels=[0,1]))

## A Confusion Matrix is a table that is often used to describe the performance
## of a classification model on a set of test data for which the true values are
## known. It allows visualization of the performance of an algorithm, typically
## a supervised learning one. Each row of the matrix represents the instances in
## an actual class while each column represents the instances in a predicted class.
## The main diagonal represents the instances that were classified correctly
## (True Positives - TP and True Negatives - TN), while off-diagonal elements
## indicate the misclassifications.
```
